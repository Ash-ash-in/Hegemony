### Trade Unions

```python
#establish trade_union class 

class trade_union:
    def __init__(self, name, skill, worker = None):
        self.__name = name # str
        self.__skill = skill # str
        self._worker = worker # classObj
        
    @property
    def skill(self):
        return self.__skill
    
    @property
    def industry(self):
        return self.__skill
    
    @property
    def worker(self):
        return self._worker
    
    def add_worker(self, workerobj):
        fail = False
        if worker:
            print('Cannot add worker to', lower(self.__name), ', slot already occupied')
            fail = True
        if workerobj.skill != self.__skill:
            print('Cannot add worker to', lower(self.__name), ', incorrect skill')
            fail = True
        if workerobj.faction != 'WC':
            print('Cannot add worker to', lower(self.__name), ', incorrect faction')
            fail = True
        if not fail:
            self._worker = workerobj
            
        
```

### Worker Setup

##### Workers are objects that frequently move and change status. As such, you should use functions related to workers, rather than companies, when it comes to assigning and unassigning them.

```python
# establish worker class

class worker:
    def __init__(self, ID, skill, faction, employer = None, in_play = False):
        self.__ID = ID # str: Unique ID
        self.__skill = skill # str: What profession is it?
        self.__faction = faction # str: What player owns this worker?
        self._employer = employer # classObj: What company is it assigned to?
        self._in_play = in_play # bool: True - on the board, False - off the board
        
    @property
    def ID(self):
        return self.__ID
    
    @property
    def skill(self):
        return self.__skill
    
    @property
    def faction(self):
        return self.__faction
    
    @property
    def employer(self):
        return self._employer
    
    @property
    def company(self):
        return self._employer
    
    @property
    def in_play(self):
        return self._in_play
    
    def assign_to_company(self, company, slotname):
        logging.info(f"Assigning {self.__ID} to {company.ID}")
        if not self._in_play:
            logging.error(f"Worker {self.__ID} not in play, cannot assign to {company.ID}")
            return False
        if self._employer:
            logging.error(f"Error assigning {self.__ID} to {company.ID}: worker already employed")
            return False
        if company.add_worker(self, slotname):
            self._employer = company
            logging.debug("Worker successfully assigned")
            return True
        else:
            logging.error("Failed to add worker to company, worker-side assignment also failed")
        
    def unassign_from_company(self, company):
        logging.debug(f"Unassigning {self.__ID} from {company.ID}")
        if self.employer != company:
            logging.error(f"Unable to unassign {self.__ID} from {company.ID}: worker is not employed here (missing in worker object)")
            return False
        if self not in company.workers:
            logging.error(f"Unable to unassign {self.__ID} from {company.ID}: worker is not employed here (missing in company object)")
            return False
        if company.remove_worker(self):
            self._employer = None
            logging.debug('Worker successfully unassigned')
            return True
        else:
            logging.error("Failed to remove worker from company, worker-side unassignment also failed")
        
    def birth_worker(self):
        logging.debug(f"Birthing worker {self.__ID}")
        if self._in_play:
            logging.error(f"Worker already in play, cannot birth")
        else:
            self._in_play = True
            logging.debug(f"Worker birthed")
            print('I SHOULD ADD TO PLAYER POP NOW')
            
    def kill_worker(self):
        logging.debug(f"Killing {self.__ID}")
        if self._employer:
            logging.error('Worker is currently employed, cannot kill')
        if not self._in_play:
            logging.error(f"Worker not in play, cannot kill")
        else:
            self._in_play = False
            logging.debug(f"Worker {self.__ID} killed")
            logging.info('I SHOULD REMOVE FROM PLAYER POP NOW')
        
    def __repr__(self):
        return f"Worker({self.__ID}, Skill: {self.__skill}, Class: {self.__faction}, Employer: {self._employer if self._employer else 'None'})"
```

### Companies

##### The only time a company should be referenced in a command is when you are doing something to that company - building it or moving it

```python
class company:
    def __init__(self, ID, owner, name, cost, product, base_production, wages = None, worker_slots = None, upgrade_value = None, upgraded = False, workers = None, company_slot = None):
        self.__ID = ID
        self._owner = owner
        self.__name = name
        self.__cost = cost
        self.__product = product
        self.__base_production = base_production
        self.__wages = wages
        self._worker_slots = worker_slots # for real use
        self.__upgrade_value = upgrade_value
        self._upgraded = upgraded
        self._workers = workers # for ease of access
        self._company_slot = company_slot
        if self._upgraded:
            self._production_amount = self.__base_production + self._upgrade_value
        else:
            self._production_amount = self.__base_production
        self._location = None
    
    @property
    def ID(self):
        return self.__ID
    
    @property
    def owner(self):
        return self._owner
        
    @property
    def name(self):
        return self.__name
        
    @property
    def cost(self):
        return self.__cost
        
    @property
    def product(self):
        return self.__product
        
    @property
    def base_production(self):
        return self.__base_production
        
    @property
    def wages(self):
        return self.__wages
        
    @property
    def worker_slots(self):
        return self._worker_slots
        
    @property
    def upgrade_value(self):
        return self.__upgrade_value
        
    @property
    def upgraded(self):
        return self._upgraded
        
    @property
    def production(self):
        return self._production_amount
        
    @property
    def production_amount(self):
        return self._production_amount
    
    @property
    def workers(self):
        return self._workers
    
    @property
    def location(self):
        return self._location
    
    def __repr__(self):
        return f"Company({self.__ID}, Name: {self.__name}, Product: {self.__product}, Location: {self._location if self._location else 'None'})"
    
    def remove_worker(self, worker):
        logging.debug(f"Removing {worker.ID} from {self.__ID} (company method)")
        # checks
        try: # throw error if we can't find that worker at this company
            self._workers.index(worker)
        except (ValueError, AttributeError):
            logging.error(f"Worker {ID(worker)} not assigned to {self.__ID}, cannot remove")
            return False
        
        # remove from worker list
        if len(self._workers) == 1:
            self._workers = None
        else:
            self._workers.remove(worker)   
        
        # remove worker from slots
        found = False
        for slotname, slot in self._worker_slots.items():
            if slot[1] == worker:
                slot[1] = None
                found = True
                break
        if not found:
            logging.error(f"Could not find worker {worker.ID} in company's worker slots. Something is broken.")
            return False
        logging.debug(f"Worker removal successful (company method complete)")
        return True
        
    def add_worker(self, worker, slotname):
        logging.debug(f"Adding {worker.ID} to {self.__ID} (company method)")
        
        #checks
        if worker.employer:
            logging.error("Error adding worker, already employed")
            return False
        try:
            self._worker_slots[slotname]
        except:
            logging.error(f"Error adding worker, slot '{slotname}' does not exist")
            return False
        try:
            if worker in self._workers:
                logging.error(f"Error adding worker, worker already employed here")
                return False
        except:
            pass
        if self._worker_slots[slotname][1]:
            logging.error(f"Error adding worker, slot '{slotname}' already occupied")
            return False
        if worker.faction != self.worker_slots[slotname][0][0] and self.worker_slots[slotname][0][0] != 'Any':
            logging.error(f"Error adding worker, '{worker.faction}' is not an appropriate faction")
            return False
        if worker.skill != self.worker_slots[slotname][0][1] and self.worker_slots[slotname][0][1] != 'Any':
            logging.error(f"Error adding worker, '{worker.skill}' is not an appropriate skill")
            return False
    
        # adding worker to list
        if not self._workers:
            self._workers = []
        self._workers.append(worker)
        logging.debug("Worker added to worker list")
        
        # adding worker to slot
        self._worker_slots[slotname][1] = worker        
        logging.debug("Worker added to slot")
            
        logging.debug("Worker successfully added (company method complete)")
        return True
    
    def establish_company(self, company_slot):
        logging.debug(f"Establishing company {self.__ID} in {company_slot} (company method)")
        if company_slots[company_slot]:
            logging.error('Error establishing company, slot already occupied')
            return False
        elif self._location:
            logging.error('Error establishing company, company already established')
            return False
        else:
            self._location = company_slot
            logging.debug('Company established')
            return True
            
            
```

```python
# create companies from csv
import pandas as pd
import os
cwd = os.getcwd()

companies_df = pd.read_csv(cwd+"\Hegemony Companies.csv")


def company_setup(faction) -> dict:
    
    if faction == 'WC':
        csv_faction = 'Working Class'
    elif faction == 'MC':
        csv_faction = 'Middle Class'
    elif faction == 'CC':
        csv_faction = 'Capitalists'
    elif faction == 'State':
        csv_faction = 'State'
    
    company_dict = {}
    company_int = 1
    for index, row in companies_df[companies_df['Owner'] == csv_faction].iterrows():
        companyID = faction +'_company' + str(company_int)

        if companyID in company_dict.keys():

            # duplicate error handling
            logging.error('Error in company setup, duplicate company created: ' + companyID)

        else: 
            # extract complex parts
            # wages
            if not pd.isna(row['L1']):
                wages = (int(row['L1']), int(row['L2']), int(row['L3']))
            else:
                wages = None
                
            # worker slots
            if pd.isna(row['Class1']):
                worker_slots = None
            else:
                worker_slots = {}
            if not pd.isna(row['Class3']):
                worker_slots['S3'] = [(row['Class1'], row['Skill1']), None]
            if not pd.isna(row['Class2']):
                worker_slots['S2'] = [(row['Class2'], row['Skill2']), None]
            if not pd.isna(row['Class1']):
                worker_slots['S1'] = [(row['Class1'], row['Skill1']), None]

            # set up object    
            company_dict[companyID] = company(
                companyID,
                row['Owner'],
                row['Name'],
                row['Cost'],
                row['Product'],
                row['Base Production'],
                wages,
                worker_slots,
                row['Upgrade Value']
            )
            company_int += 1
            
    return company_dict
        
# ID, owner, name, cost, product, base_production, wages = None, worker_slots = None, upgrade_value = None, upgraded = False, workers = None, company_slot = None):

company_pool = {
    'WC_companies':company_setup('WC'), 
    'MC_companies':company_setup('MC'), 
    'CC_companies':company_setup('CC'), 
    'State_companies':company_setup('State')
}

logging.debug('company_pool created')
```

```python
def find_worker(faction, skill, in_play = 'Yes', exceptions = [], company = None) -> object:
    """
    faction: which class' workers are you looking for?
    skill: required skill, 'Any' if it doesn't matter
    in_play: 'No' for finding off-board workers, 'Any' is for setup only    
    company: enter a company object to restrict search to those employed at a specific place. None will search those without a company
    exceptions: list of workers to ignore in the search
    """

    logging.debug(f"Searching for worker with criteria: faction = {faction}, skill = {skill}, in_play = {in_play}, company = {company if company else 'None'}")
    
    if faction == 'WC':
        search_dict = 'WC_workers'
    elif faction == 'MC':
        search_dict = 'MC_workers'
    else:
        search_dict = 'Both'
    
    def search_loop(active_dict) -> list:
        
        eligible_workers = []
        for ID, worker in worker_pool[active_dict].items():
            
            # check exceptions
            if worker in exceptions:
                #logging.debug(f"Worker {worker} rejected due to exception rule")
                continue
            
            # check skill requirement
            if skill != 'Any' and worker.skill != skill:
                #logging.debug(f"Worker {worker} rejected due to skill requirement")
                continue
            
            # check in play requirement
            if in_play == 'Yes' and not worker.in_play:
                #logging.debug(f"Worker {worker} rejected due to in play requirement")
                continue
            elif in_play == 'No' and worker.in_play:
                #logging.debug(f"Worker {worker} rejected due to in play requirement")
                continue
            
            # check company requirement
            if company and worker.employer != company:
                #logging.debug(f"Worker {worker} rejected due to company requirement")
                continue
            elif not company and worker.employer:
                #logging.debug(f"Worker {worker} rejected due to company requirement")
                continue
            
            # add anyone left
            #logging.debug(f"Worker {worker} accepted")
            eligible_workers.append(worker)
            
        return eligible_workers
                
    # execute loop based on worker faction requirement
    if search_dict != 'Both':
        eligible_workers = search_loop(search_dict)   
    else:
        eligible_workers = search_loop('WC_workers') + search_loop('MC_workers')
        
    logging.debug(f"Found {len(eligible_workers)} eligible workers")
    
    # handling if there are no eligible candidates
    if len(eligible_workers) == 0:
        logging.info('Failed to find any suitable candidates')
        return
    
    # dealing with 'any' skill - prioritise unskilled
    elif skill == 'Any':
        for candidate in eligible_workers:
            if candidate.skill == 'Unskilled':
                logging.debug(f"Worker search returning {candidate}")
                return candidate
            
    # return first worker in list
    logging.debug(f"Worker search returning {eligible_workers[0]}")
    return eligible_workers[0] 
                      
```

### Company Slots

```python
#set up company slot dict
company_slots = {}
for i in range(12):
    company_slots['CC_company_slot' + str(i + 1)] = None
    company_slots['State_company_slot' + str(i + 1)] = None
    if i < 2:
        company_slots['WC_company_slot' + str(i + 1)] = None
    if i < 8:
        company_slots['MC_company_slot' + str(i + 1)] = None    
        
# set up assignment function
def assign_company_to_slot(company, slotname):
    
    global company_slots
    logging.debug(f"Assigning company {company.ID} to {slotname}")
    if company.establish_company(slotname):
        company_slots[slotname] = company
        logging.info('Assignment successful')
        return True
    else:
        logging.error('Failed to assign in company method. Failing function.')
        return False
    
```

```python
def setup_companies(player_count):

    logging.info('Starting company setup')
    global company_slots
    
    # set up capitalist companies
    CC_setup = ['Supermarket', 'Shopping Mall', 'College', 'Clinic']
    for company in company_pool['CC_companies'].values():
        if company.name in CC_setup:
            logging.debug(f"Setting up {company.name}")
            slotname = 'CC_company_slot' + str(5 - len(CC_setup))
            assign_company_to_slot(company, slotname)
            CC_setup.remove(company.name)

    # set up state companies
    if player_count == 2:
        State_setup = ['Regional TV Station', 'Public Hospital', 'Public University']
    elif player_count > 2:
        State_setup = ['Technical University', 'University Hospital', 'National Public Broadcasting']
    for company in company_pool['State_companies'].values():
        if company.name in State_setup:
            logging.debug(f"Setting up {company.name}")
            slotname = 'State_company_slot' + str(4 - len(State_setup))
            assign_company_to_slot(company, slotname)
            State_setup.remove(company.name)    

    # setup middle class companies
    if player_count > 2:
        MC_setup = ['Convenience Store', "Doctor's Office"]
        for company in company_pool['MC_companies'].values():
            if company.name in MC_setup:
                logging.debug(f"Setting up {company.name}")
                slotname = 'MC_company_slot' + str(3 - len(MC_setup))
                assign_company_to_slot(company, slotname)
                MC_setup.remove(company.name)    

    logging.info('Company Setup Complete')
    return

setup_companies(player_count)
```

```python
# action functions - worker placement

def staff_from_abroad(company, faction):
    logging.debug(f"Staffing {company} from abroad")
    
    # check all slots have available workers
    Failure = False
    candidates = {}
    for key, value in company.worker_slots.items():
        logging.debug(f"Finding slot {key}")
        try:
            worker = find_worker(faction, value[0][1], 'No', list(candidates.values()))
        except:
            Failure = True
        else:
            candidates[key] = worker
                      
    if not Failure:
        logging.debug('All worker slots found, staffing company')
        for slot, worker in candidates.items():
            worker.birth_worker()
            worker.assign_to_company(company, slot)
        return True
                      
    else:
        logging.info('Unable to fill all slots of the company. Terminating staffing attempt')
        return False
```

```python
# setup immigration cards
immigration_cards = []
for union_name in trade_unions:
    for i in range(3):
        if i < 2:
            immigration_cards.append({'WC':union_name.industry, 'MC':'Unskilled'})
        immigration_cards.append({'MC':union_name.industry, 'WC':'Unskilled'})
        
random.shuffle(immigration_cards)

def draw_immigration_card(faction):
    
    # select first card, and put it to the back of the deck
    card = immigration_cards[0]
    immigration_cards.remove(card)
    immigration_cards.append(card)
    
    # move worker to in-play
    worker = find_worker(faction, card[faction], 'No')
    worker.birth_worker()
    return
    
```

```python
def setup_workers(player_count):

    ### Working Class
    
    # define companies to assign workers to (dependant on player count) and draw immigration cards
    if player_count == 2:
        WC_setup = ['Supermarket', 'Shopping Mall' ,'Public Hospital', 'Public University']
        draw_immigration_card('WC')
        
    else:
        WC_setup = ['Supermarket', 'College', 'University Hospital']
        draw_immigration_card('WC')
        draw_immigration_card('WC')

    # assign workers to companies
    for company in company_slots.values():
        if not company:
            continue
        if company.name in WC_setup:
            logging.debug(f"Executing worker setup for {company.name}")
            staff_from_abroad(company, 'WC')
            WC_setup.remove(company.name)
            
    # spawn one worker in unemployment area        
    find_worker('WC', 'Any', 'No').birth_worker()   


    ### Middle Class
    if player_count > 2:
        MC_setup = ['Technical University', 'Shopping Mall'] 
        MC_setup2 = ['Convenience Store', "Doctor's Office"]
        for company in company_slots.values():
            if not company:
                continue
            if company.name in MC_setup:
                logging.debug(f"Executing worker setup for {company.name}")
                staff_from_abroad(company, 'MC')
                MC_setup.remove(company.name)
            elif company.name in MC_setup2:
                logging.debug(f"Executing worker setup for {company.name}")
                worker = find_worker('MC', company.worker_slots['S1'][0][1], 'No')
                worker.birth_worker()
                worker.assign_to_company(company, 'S1')
                MC_setup2.remove(company.name)
                
        skill = str(input('Middle Class Player: pick a skilled worker to add to the unemployment area')).strip().capitalize()
        while skill not in ['Agriculture', 'Luxury', 'Healthcare', 'Education', 'Media']:
            skill = str(input('Input not understood, try again')).strip().capitalize()
        find_worker('MC', skill, 'No').birth_worker()
        
        draw_immigration_card('WC')
        draw_immigration_card('WC')
                   
    return

setup_workers(player_count)
            
```

```python
class Player():
    
    def __init__(self, faction, income, resources, victory_points, loans):
        self.__faction = faction
        self._income = income
        self._resources = resources
        self._victory_points = victory_points
        self._loans = loans
        
    def add_victory_points(self, amount):
        logging.debug(f"Adding {amount} victory points to {faction}")
        self._victory_points += amount
        
    def receive_money(self, amount):
        logging.debug(f"Adding ${amount} to {faction}")
        if amount <= 0:
            logging.error('Failed to add money - amount cannot be less than 0')
            return False
        else:
            self._income += amount
            
    def pay_money(self, amount, mandatory = False):
        logging.debug(f"Removing ${amount} from {faction}")
        if amount > self._income:
            self._income -= amount
        elif not mandatory:
            logging.debug('Not enough money to make payment')
            return False
        elif mandatory:
            logging.debug('Not enough money for mandatory payment, loan required')
            self._income -= amount
            while self._income < 0:
                self.take_loan()
            
    def take_loan(self):
        logging.debug(f"{faction} taking a loan")
        self._loans += 1
        self.receive_money(50)
        
class WorkingClass(Player):
    
    def __init__(self, faction, money, resources, victory_points, loans, worker_count, prosperity, resources):
        super().__init__(faction, money, resources, victory_points, loans)
        self._prosperity = prosperity
        self._worker_count = worker_count
        self._resources = resources
        self._population = population_calc(worker_count)
        
    def population_calc(worker_count):
        logging.debug(f"Recalculating population for {self.__faction})
        if worker_count < 30 and worker_count > 9:
            population = worker_count // 3
        elif worker_count >= 30:
            population = 10
        else:
            population = 3
        return population
                      
    def add_prosperity():
        logging.debug('Adding WC prosperity')      
        if self._prosperity < 10:
            self._prosperity += 1
        self.add_victory_points(self._prosperity)
                      
    def drop_prosperity():
        logging.debug('Dropping WC prosperity')     
        if self._prosperity < 0:
            self._prosperity -= 1
    
class MiddleClass(WorkingClass):
    
    def __init__(self, faction, money, resources, victory_points, loans, worker_count, prosperity, resources, trade_goods, prices, storage):
        super().__init__(faction, money, resources, victory_points, loans, worker_count, prosperity_position, resources)   
        self._prosperity = prosperity
        self._resources = resources
        self._population = population_calc(worker_count)
        self._trade_goods = trade_goods
        self._prices = prices
        self._storage = storage
        
    prosperity_tracker = [0,1,2,3,4,5,5,6,6,7,7]
        
    def add_prosperity():
        logging.debug('Adding MC prosperity')
        if self._prosperity_position + amount > 10:
            self._prosperity_position = 10
        else:
            self._prosperity_position += 1
        self._prosperity = prosperity_tracker[self._prosperity_position]
        self.add_victory_points(self._prosperity)
        
    def drop_prosperity():
        logging.debug('Dropping MC prosperity')
        if self._prosperity_position > 0:
            self._prosperity_position -= 1
        self._prosperity = prosperity_tracker[self._prosperity_position]
        
class CapitalistClass(Player):
                      
    def __init__(self, faction, money, resources, victory_points, trade_goods, prices, storage):
        super().__init__(faction, money, resources, victory_points)
        self._trade_goods = trade_goods
        self._prices = prices              
        self._storage = storage
```
